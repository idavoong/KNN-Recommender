```python
import duckdb
```

```python
# Content Views
#  content_id  seconds_viewed  rating  adventurer_id  publisher_id  playlist_id   month     day     day_of_month  year  
#   varchar        double      double     varchar       varchar       varchar    varchar  varchar      int64      int64 

cols = ["content_id", "adventurer_id", "publisher_id", "playlist_id", "month", "day"," day_of_month", "year"]
cols2 = ["seconds_viewed", "rating"]

for col in cols:
    values = duckdb.sql(f"""
        SELECT DISTINCT {col}
        FROM "data/content_views.parquet"
        ORDER BY {col} ASC
    """)
    print(f"Unique values in column '{col}': {[row[0] for row in values.fetchall()]}")

for col in cols2:
    ranges = duckdb.sql(f"""    
        SELECT MIN({col}), MAX({col})
        FROM "data/content_views.parquet"
    """)
    print(f"Range for '{col}': {ranges.fetchall()}")
```

```python
# Content Metadata
#  content_id  studio   title   genre_id  minutes  language_code    month       day     day_of_month  year  
#   varchar    varchar varchar  varchar   double      varchar      varchar    varchar      int64      int64 

cols = ["content_id", "studio", "title", "genre_id", "language_code", "month", "day"," day_of_month", "year"]

for col in cols:
    values = duckdb.sql(f"""
        SELECT DISTINCT {col}
        FROM "data/content_metadata.parquet"
        ORDER BY {col} ASC
    """)
    print(f"Unique values in column '{col}': {[row[0] for row in values.fetchall()]}")

ranges = duckdb.sql(f"""    
    SELECT MIN(minutes), MAX(minutes)
    FROM "data/content_metadata.parquet"
""")
print(f"Range for 'minutes': {ranges.fetchall()}")
```
